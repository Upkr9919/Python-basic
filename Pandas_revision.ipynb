```python
import pandas as pd

# Load your dataset
df = pd.read_csv('/content/pokemon.csv')
df
```

```python
# Detect column names (handling case sensitivity)
type_col = 'Type 1' if 'Type 1' in df.columns else 'type1'
name_col = 'Name' if 'Name' in df.columns else 'name'
speed_col = 'Speed' if 'Speed' in df.columns else 'speed'

# FIX: Removed .squeeze("columns") because selecting a single column
# from df.set_index() already returns a clean Series.
poke_types = df.set_index(type_col)[name_col]
poke_speed = df.set_index(name_col)[speed_col]
```

# **# Challenge 1: The Speed Demon (Filtering & Indexing)**
Dataset: `poke_speed`
### **Task:**
1.   Find the average speed of all Pokémon.
2.   Filter the Series to find all Pokémon that are at least twice as fast as the average speed.
3.   Sort these fast Pokémon in descending order and return just the top 5 fastest names.

```python
print(poke_speed.mean())
```

```python
print(poke_speed[poke_speed >= 2*(poke_speed.mean())])
```

```python
print(poke_speed[poke_speed >= 2*(poke_speed.mean())].sort_values(ascending=False).head(5))
```

## Challenge 3: Advanced Binning & Analysis

**Dataset:** `poke_speed`  
**Tasks:**
1. Group the Pokémon speeds into 3 distinct categories: "Slow", "Medium", and "Fast" using `pd.cut()`.
2. Convert this new Series into a categorical type.
3. Calculate the exact percentage of Pokémon that fall into the "Slow" category.


```python
poke_speed
```

```python
speed = pd.cut(poke_speed, bins=3, labels=['Slow', 'Medium', 'Fast'])

speed
```

```python
speed.value_counts(normalize=True)
```

```python
a = speed.value_counts()[0]

b = speed.value_counts().sum()

c = a/b*100

print(round(c,2))
```

```python
print(speed.value_counts(normalize=True)[0] * 100)
```
